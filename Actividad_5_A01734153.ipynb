# **California_EUA**

## **2. Agregar el archivo csv**

```python
california= pd.read_csv("California_EUA.csv",encoding = "unicode_escape")
california.info()
```

## **3. Reemplazar valores nulos**

```python
#Identificar valores nulos por columna (poner opción de mostrar todas las columnas)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    valores_nulos = california.isnull().sum()
    print(valores_nulos)
```

```python
#Se separan las columnas numéricas de las que contienen otro tipo de dato. 
#Columnas con str, date_time, etc
columnas_object = california.select_dtypes(exclude=['float', 'int'])

#Columnas numéricas
columnas_numericas = california.select_dtypes(include=['float', 'int'])
```

### **Columnas no numéricas**

Para las columnas que no tienen valores numéricos se hará un filtrado donde se sustituirán los valores nulos por el str "Sin registro". Esto tiene el objetivo de generar insights a partir de los datos correctos pero también identificar el porcentaje de filas que no tienen registro para ciertas columnas. Además evita no borrar las columnas que tienen completamente valores nulos 

```python
#Comprobamos nulos para las columnas no numéricas
columnas_object.isnull().sum()
```

```python
#LLenar los nulos con Sin registro y posteriormente guardarlo como df
columnas_object.fillna('No Data Available', inplace=True)
```

```python
#Se guarda en el nuevo Data Frame
california_str = pd.DataFrame(columnas_object)
california_str.isnull().sum()
```

### **Columnas numéricas**

Para las columnas numéricas y debido a la naturaleza de los datos se eligió sustituir los valores nulos por la mediana de la columna 

```python
#Comprobamos nulos para las columnas numéricas
columnas_numericas.isnull().sum()
```

```python
#LLenar los nulos con la mediana de las columnas
columnas_numericas=columnas_numericas.fillna(columnas_numericas.median())
```

```python
#Se guarda en el nuevo Data Frame y se comprueba
california_num = pd.DataFrame(columnas_numericas)
california_num.isnull().sum()
```

```python
#Las columnas que tenían todos los valores nulos se mantuvieron igual
#Por este motivo se hace un nuevo procedimiento para eliminar las 3 nulas
california_num=california_num.dropna(axis=1)
california_num.isnull().sum()
```

### **Separar variables no categóricas**
Se separaran todas las columnas numéricas que tienen valores únicos, por ejemplo el ID, etc. Estas columnas no son categóricas y no representan importancia en el análisis para outliers

```python
california_num.nunique()
```

```python
#Se separan columnas numéricas con valores unicos
#Este nuevo df lo concatenaremos más adelante

num_valores_unicos = california_num.nunique()

# Crear un nuevo dataframe que almacene las variables con muchos valores unicos
columnas_muchos = num_valores_unicos[num_valores_unicos > 3000].index
california_muchos = california_num.loc[:, columnas_muchos]

# Crear un nuevo dataframe que excluya las columnas con muchos valores únicos
columnas_analisis = num_valores_unicos[num_valores_unicos < 3000].index
california_num = california_num.loc[:, columnas_analisis]
california_num
```

```python
california_muchos
```

## **4. Eliminación de Outliers**
Aplicar métodos (Desviación Estándar o Rango Intercuartílico)

Se decidió utilizar el método de Rango Intercuartílico debido a su fiabilidad y mejor precisión para detectar outliers. El método es mucho má compacto para detectar límites superiores e inferiores, y al ser archivos grandes, siempre es mejor detectar con un método más preciso

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=california_num

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print()
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= california_num[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
california_iqr= california_num[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
california_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=california_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
california_iqr_final=california_iqr.copy()
california_iqr_final=california_iqr_final.fillna(round(california_iqr.mean(),1))
california_iqr_final
```

```python
#Corroboramos valores nulos del dataframe limpio
valores_nulos=california_iqr_final.isnull().sum()
valores_nulos
```

```python
#Finalmente concatenamos el dataframe numerico con el de string y guardamos en Excel 
california_final = pd.concat([california_muchos,california_iqr_final, california_str], axis=1)
california_final.to_csv('California_EUA_LIMPIO.csv', index=False)
```

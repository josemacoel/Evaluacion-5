# **California_EUA**

```python
california= pd.read_csv("California_EUA_LIMPIO.csv",encoding = "unicode_escape")
california.info()
```

## **a) host_acceptance_rate>50%**

Primero se realizó una limpieza de los datos, reemplazando los valores faltantes por 5%. Posteriormente se reemplazaron los valores tipo string y finalmente se realizó el filtro. Se puede observar que una mayoría de los datos coinciden con el criterio, siendo 5679 filas de 6936

```python
california["host_acceptance_rate"] = california["host_acceptance_rate"].replace("No Data Available", "5%")
california['host_acceptance_rate'] = pd.to_numeric(california['host_acceptance_rate'].str.replace('%', ''))
acceptance = california[california["host_acceptance_rate"] > 50]
acceptance
acceptance.to_csv('filtroa_usa.csv', index=False)
```

## **b) Sólo los registros con categoría super host**

En este codigo se realizó la búsqueda de aquellos valores con categoría super host, que estaban definidos con una t de true en la columna del propio nombre

```python
super_host = california.loc[california['host_is_superhost'] == 't']
super_host
super_host.to_csv('filtrob_usa.csv', index=False)
```

## **c) Los registros que no hallan verificado identidad “ not identity_verified”**


De la misma forma que el código anterior, se realizó un filtro que buscara aquellos registros que no tenian idetidad verificada. 

```python
not_identify = california.loc[california['host_identity_verified'] == 'f']
not_identify
not_identify.to_csv('filtroc_usa.csv', index=False)
```

## **d) Los registros cuyo property_type = “Private room” y “Hotel room”**


En materia de filtros más específicos encontramos este, con un resultado de 4 filas que coinciden con la búsqueda de private room y hotel room. 

```python
property_type = california.loc[(california['property_type'] == 'Private room') | (california['property_type'] == 'Hotel room')]
property_type
property_type.to_csv('filtrod_usa.csv', index=False)
```

## **e) Los registros que cuenten con bathroom > 1**


Al ser una columna en formato string este codigo fue más complejo. Primero se realizó la función unique() para mostrar los valores y así identificar cual texto equivalía a 1 o 0.5. Posteriormente se realizaron dos métodos, primero reemplazar los datos escritos por números en string y finalmente extraer dentro del string los dígitos. 

```python
california["bathrooms_text"].unique()
```

```python
#cambiar Half-bath, Shared half-bath, Private half-bath, sin registro
california["bathrooms_text"] = california["bathrooms_text"].str.replace('Half-bath', '0.5').str.replace('Shared half-bath', '0.5').str.replace('Private half-bath', '0.5').str.replace('No Data Available', '0')

california["bathrooms_text"] = california["bathrooms_text"].str.extract('(\d+\.?\d*)', expand=False)
california["bathrooms_text"] = california["bathrooms_text"].astype(float)

bathroom = california[california["bathrooms_text"] > 1]
bathroom
bathroom.to_csv('filtroe_usa.csv', index=False)
```

## **f) Los registros cuyo precio sea mayor de $10,000 y que sean de tipo “Entire home”**


Para este ejercicio de filtrado se encontraron 2 variables que coincidían con un precio mayor a $10,000 dólares y que fueran Entire Home. Se puede afirmar que son las casas más caras de este tipo. 

```python
california['price'] = california['price'].str.replace(',', '').str.replace('$', '').astype(float)
price = california.loc[(california['price'] > 10000) & (california['room_type'] == 'Entire home/apt')]
price
price.to_csv('filtrof_usa.csv', index=False)
```

## **g) Los registros cuyo review_scores_cleanliness > 4.5**


Se encontró que una gran mayoría de los registros cumple con un review mayor a 4.5 para el score de cleanliness

```python
review= california[california['review_scores_cleanliness'] > 4.5]
review
review.to_csv('filtrog_usa.csv', index=False)
```

## **h) Los registros cuyo review_scores_value > 4.9**


A partir de este ejercicio de filtrado se puede observar que muy pocos registros logran la excelencia en esta categoría de score. 

```python
review_scores= california[california['review_scores_value'] > 4.9]
review_scores
review_scores.to_csv('filtroh_usa.csv', index=False)
```

## **i) Los registros cuya availability_365< 100**

Se filtaron aquellos registros cuya disponibilidad anual es menor a 100 días

```python
availability= california[california['availability_365'] < 100]
availability
availability.to_csv('filtroi_usa.csv', index=False)
```

## **j) Los registros cuya host_response_time sea “within an hour”**

Un porcentaje significativo de host responden dentro del marco de una hora, obteniendo 4221 filas de este filtro. 

```python
response= california[california['host_response_time'] == 'within an hour']
response
response.to_csv('filtroj_usa.csv', index=False)
```
